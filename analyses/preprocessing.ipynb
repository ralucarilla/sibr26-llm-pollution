```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df1 = pd.read_csv('./data/raw/ai_fat_wave3_pt1.csv')
df2 = pd.read_csv('./data/raw/ai_fat_wave3_pt2.csv')

# add a column indicating the part
df1['part'] = 'part_1'
df2['part'] = 'part_2'
```

```python
df_demographics1 = pd.read_csv('./data/raw/prolific_demographic_export_68e4ce20a1b531e5b6bfaf1b.csv') 
df_demographics2 = pd.read_csv('./data/raw/prolific_demographic_export_68e4cebf0dfea69a27122b30.csv') 
```

```python
# combine the two dataframes
df = pd.concat([df1, df2], ignore_index=True)
df_demographics = pd.concat([df_demographics1, df_demographics2], ignore_index=True)
```

```python
# match by participant_label to add all columns from df_demographics to df
df = df.merge(df_demographics, left_on='participant.label', right_on='Participant id', how='left')
df.shape
```

```python
df.head()
```

```python
# drop rows where partcipant._current_page_name is not 'p13_thanks'
df = df[df['participant._current_page_name'] == 'p13_thanks']
```

```python
# remove columns that are not needed
cols_to_remove = ['participant.id_in_session','participant._is_bot', 'participant._index_in_pages',
       'participant._max_page_index', 'participant._current_app_name','participant.visited', 'participant.mturk_worker_id',
       'participant.mturk_assignment_id', 'participant.payoff', 'session.code', 'session.label',
       'session.mturk_HITId', 'session.mturk_HITGroupId', 'session.comment',
       'session.is_demo', 'session.config.real_world_currency_per_point',
       'session.config.prolific_completion_link',
       'session.config.recaptcha_site_key',
       'session.config.prolific_no_consent_link', 'session.config.name',
       'session.config.recaptcha_secret_key',
       'session.config.participation_fee', '_1_consent.1.player.id_in_group',
       '_1_consent.1.player.role', '_1_consent.1.player.payoff', '_1_consent.1.group.id_in_subsession',
       '_1_consent.1.subsession.round_number', '_2_main.1.player.id_in_group',
       '_2_main.1.player.role', '_2_main.1.player.payoff', '_2_main.1.group.id_in_subsession',
       '_2_main.1.subsession.round_number',
       '_2_main.1.subsession.prolific_completion_link',
       '_2_main.1.subsession.prolific_no_consent_link',
       '_3_exit_questions.1.player.id_in_group',
       '_3_exit_questions.1.player.role', '_3_exit_questions.1.player.payoff', '_3_exit_questions.1.group.id_in_subsession',
       '_3_exit_questions.1.subsession.round_number',
       '_4_demographics.1.player.id_in_group', '_4_demographics.1.player.role',
       '_4_demographics.1.player.payoff', '_4_demographics.1.group.id_in_subsession',
       '_4_demographics.1.subsession.round_number',
       '_5_feedback.1.player.id_in_group', '_5_feedback.1.player.role',
       '_5_feedback.1.player.payoff', '_5_feedback.1.player.recaptcha_token', '_5_feedback.1.group.id_in_subsession',
       '_5_feedback.1.subsession.round_number',
       'Participant id', 'Submission id', 'Custom study tncs accepted at',
       'Completion code', 'Archived at', 'Started at', 'Completed at',
       'Reviewed at', 'Status', 'Nationality', 'Language', 'participant.label']
df = df.drop(columns=cols_to_remove)
```

```python
# rename columns to keep only last part after last dot
df.columns = [col.split('.')[-1] for col in df.columns]
```

```python
df.columns
```

```python
# rename vars
df = df.rename(columns={'check': 'honeypot_checkbox'})
df = df.rename(columns={'Age': 'age_prolific'})
df = df.rename(columns={'Ethnicity simplified': 'ethnicity'})
df = df.rename(columns={'Gender': 'gender'})
df = df.rename(columns={'Country of birth': 'country_birth'})
df = df.rename(columns={'Country of residence': 'country'})
df = df.rename(columns={'Student status': 'student'})
df = df.rename(columns={'Employment status': 'employment'})
```

```python
for i in range(1, 6):
    df[f'association{i}_lower'] = df[f'association{i}'].str.lower()
```

```python
# rename _current_page_name to finished and set all values to True
df = df.rename(columns={'_current_page_name': 'finished'})
df['finished'] = True
```

```python
# save to new csv
df.to_csv('./data/clean/ai_fat_wave3.csv', index=False)
```

```python
all_cols = df.columns.tolist()
stub_cols = [f'association{i}' for i in range(1, 6)]
timer_cols = [f'association{i}_timer' for i in range(1, 6)]

other_cols = [col for col in all_cols if col not in stub_cols and col not in timer_cols]

df_long = pd.wide_to_long(
    df,
    stubnames=['association', 'association_timer'],
    i=other_cols,                              
    j='association_number'                    
).reset_index()

df_long['association'] = df_long['association']
df_long['association_lower'] = df_long['association'].str.lower()
```

```python
timer_col_index = df_long.columns.get_loc('association_timer')
df_long.insert(
    timer_col_index, 
    'association_lower', 
    df_long.pop('association_lower')
)

df_long = df_long.drop(columns=['association_number'])
```

```python
df_long.to_csv('./data/clean/ai_fat_wave3_long.csv', index=False)
```
